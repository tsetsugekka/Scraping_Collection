```python
df2 = pd.read_csv('20210521社内データ（更新分）_step2.csv')

browser = webdriver.Chrome()

browser.get("https://8card.net/myhome")
browser.add_cookie({'name':'_8sesp_','value':cookies['_8sesp_']})
browser.add_cookie({'name':'__stripe_mid','value':cookies['__stripe_mid']})
browser.add_cookie({'name':'_ga','value':cookies['_ga']})
browser.add_cookie({'name':'_gcl_au','value':cookies['_gcl_au']})
browser.add_cookie({'name':'_gid','value':cookies['_gid']})
browser.add_cookie({'name':'_ts_yjad','value':cookies['_ts_yjad']})
browser.add_cookie({'name':'autologin','value':cookies['autologin']})
browser.get("https://8card.net/myhome")

num=0
err=[]
ots,mts,ems,cns,dns,pns,uids=[],[],[],[],[],[],[]

df3=df2
for index,row in df3.iterrows():
    browser.get(row["eight URL"])
    random_num(2)
    
    uid=row["eight URL"].split("/")[-1]
    
    try:
        browser.find_element_by_link_text('TOPに戻る')
        office_tel,mobile_tel,mail,company,department,postion="","","","","",""
    except:
        try:
            browser.find_element_by_xpath('//*[@class="common-text_button more-info-button"]').click()
        except:
            pass
        time.sleep(1)

        tels_in_page=browser.find_elements_by_xpath('//*[@class="text-block tel"]')
        #オフィス電話
        office_tel=""
        try:
            office_tel_temp=tels_in_page[0].text
            if office_tel_temp[0:3] not in ["050","070","080","090"]:
                office_tel=office_tel_temp
        except:
            pass

        #携帯電話
        mobile_tel=""
        for i in tels_in_page:
            i=i.text
            if i[0:3] in ["050","070","080","090"]:
                mobile_tel=i

        #メール
        try:
            mail=browser.find_element_by_xpath('//*[@rel="nofollow noopener noreferrer"]').text
        except:
            mail=""

        #その他
        etc=browser.find_elements_by_xpath('//*[@class="text-block"]')
        try:
            company=etc[0].text
        except:
            company=""
        
        try:
            department=etc[1].text
        except:
            department=""

        try:
            check_str=etc[2].text
            if check_str.isdigit():
                postion=""
            else:
                postion=check_str
        except:
            postion=""
    
    ot,mt,em,cn,dn,pn=office_tel,mobile_tel,mail,company,department,postion
    if cn=="":
        err.append(index)
    else:
        ots.append(ot)
        mts.append(mt)
        ems.append(em)
        cns.append(cn)
        dns.append(dn)
        pns.append(pn)
        uids.append(uid)
    
    random_num(4)

    #print(index,"\n",company,"\n",department,"\n",postion)
        
    #カウント処理
    num=num+1
    if num%100==0:
        print("\r現在{0}件目".format(num), end="")
        
browser.close()

indexlist=list(df3.index)

for e in err:
    indexlist.remove(e)

df3.loc[indexlist,"電話番号"]=ots
df3.loc[indexlist,"携帯電話"]=mts
df3.loc[indexlist,"E-mail"]=ems
df3.loc[indexlist,"所在企業"]=cns
df3.loc[indexlist,"部署"]=dns
df3.loc[indexlist,"役職"]=pns
df3.loc[indexlist,"更新時間"]=timestamp
df3.loc[indexlist,"uid"]=uids
df3.loc[indexlist,"社内パス有無"]=str(1)
df3=df3.drop(columns=["部署役職"])

df3 = df3[["社内パス有無","更新時間","名前","所在企業","部署","役職","社内パス","電話番号","携帯電話","E-mail","eight URL","uid"]]

df3.to_csv(timestamp+"社内データ（更新分）_step3.csv", index=False)
df3
```
