```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import Select
from datetime import datetime
import time,re
import pandas as pd
import chromedriver_binary
import random
def random_num(max_num):
    return random.randint(1,max_num)
```

```python
browser = webdriver.Chrome()
browser.get("https://paypay.ne.jp/shop/")

#adを除く
browser.find_element_by_xpath('/html/body/div[1]/div[2]/div/p').click()

#地域リスト作成
arealist=browser.find_elements_by_xpath('//a[@class="regionTab__link"]')

areacontain=[]

for area in arealist:
    a=area.text
    u=area.get_attribute('href')
    areacontain.append([a,u])

#地域別ページのスクレイピング

contents=[]

for areaset in areacontain:
    
    #areaset 0地域  1URL
    browser.get(areaset[1])
    
    cat=browser.find_elements_by_xpath('//h3[@class="available__category"]')
    name=browser.find_elements_by_xpath('//ul[@class="available__storeList storeList"]')
    
    #c　カテゴリー　　n 店舗名
    for c,n in zip(cat,name):
        c=c.text
        nl=n.text.split("\n")
        for i in nl:
            contents.append([areaset[0],c,i])
            
df=pd.DataFrame(contents,columns=["地域","カテゴリー","店舗"])
df.to_excel('20201016_PayPay加盟店スクレイピング.xlsx', sheet_name='sheet1',index=False)
df
```
