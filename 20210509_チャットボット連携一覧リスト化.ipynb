```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import Select
from datetime import datetime,date
import time,re
import pandas as pd
import chromedriver_binary
import random
def random_num(max_num):
    return random.randint(2,max_num)
import json
```

```python
url="https://assistant.google.com/explore/search?q=%E3%81%A4%E3%81%AA%E3%81%84%E3%81%A7&hl=ja"

browser = webdriver.Chrome()
browser.get(url)
time.sleep(30) 

ttls=browser.find_elements_by_xpath('//div[@class="FdWgBb nhLye"]')
dtls=browser.find_elements_by_xpath('//div[@class="vsao6c-ppR2Cb CdFZQ"]')
urls=browser.find_elements_by_xpath('//div[@class="SNU6J"]/a')

print(len(ttls))
delta=[]
for t,d,u in zip(ttls,dtls,urls):
    t=t.text
    d=d.text
    u=u.get_attribute("href")
    delta.append([t,d,u])
    
df1=pd.DataFrame(delta,
            columns=['機能', '概要', 'URL'])
df1
```

```python
df=df1
browser = webdriver.Chrome()

lm,ld,lt,lc,ltr,lca,lr,lu=[],[],[],[],[],[],[],[]

for index,raw in df.iterrows():    
    browser.get(raw['URL'])
    msgs=browser.find_elements_by_xpath('//div[@class="FKHNF AdBzde"]')
    msgl=["「"+msg.text+"」" for msg in msgs]
    m="".join(msgl)
    detail=browser.find_element_by_xpath('//div[@class="IB9ccf"]').text
    ttl=browser.find_element_by_xpath('//div[@class="YtWsM RfR9R"]').text
    cpn=browser.find_element_by_xpath('//div[@class="lUcxUb CbqDob"]').text
    categorys=browser.find_elements_by_xpath('//div[@class="rkJR4e CdFZQ"]')
    category=categorys[0].text
    rank=browser.find_element_by_xpath('//div[@class="NRNQAb FTlnHb"]').text
    user=browser.find_element_by_xpath('//div[@class="rriIab CdFZQ"]').text

    if  "Google" in cpn:
        trd="自社"
    else:
        trd="他社"

    lm.append(m)
    ld.append(detail)
    lt.append(ttl)
    lc.append(cpn)
    ltr.append(trd)
    lca.append(category)
    lr.append(rank)
    lu.append(user)
    time.sleep(random_num(2))

browser.close()

df["呼び方"]=lm
df["説明"]=ld
df["会社"]=lc
df["自他社"]=ltr
df["カテゴリー"]=lca
df["評価"]=lr
df["評価ユーザー数"]=lu
df.to_excel("20210510_Google連携サービス.xlsx", encoding='utf-8', index=False)
df
```
