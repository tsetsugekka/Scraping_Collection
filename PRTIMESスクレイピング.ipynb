```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import Select
from datetime import datetime
import time,re
import pandas as pd
import chromedriver_binary
import random
def random_num(max_num):
    return random.randint(1,max_num)
```

```python
#企業URLの取得

#DBログイン
browser = webdriver.Chrome()
browser.get("https://search.yahoo.co.jp/")
browser.maximize_window()

#要検索企業リストを導入
df2 = pd.read_excel('sample2_10.xlsx')
df2=df2.reset_index()

keyword2=" site:https://prtimes.jp/"
database=[]

for index, row in df2.iterrows():
    keyword1=row["DB会社名"]
    keyword1=keyword1.translate(str.maketrans({chr(0xFF01 + i): chr(0x21 + i) for i in range(94)})) #全角 -> 半角
    keyword="\""+keyword1+"のプレスリリース\""+keyword2
    no=index

    browser.get("https://search.yahoo.co.jp/")
    browser.find_element_by_xpath('//*[@id="yschsp"]').send_keys(keyword)
    browser.find_element_by_xpath('//*[@id="sbn"]/fieldset/div[1]/input').click()

    try:
        title=browser.find_element_by_xpath('//*[@id="js-tracked_mod_4"]/div/section/div[1]/div/div/a[1]/h3').text
        url=browser.find_element_by_xpath('//*[@id="js-tracked_mod_4"]/div/section/div[1]/div/div/a[1]/div/cite').text
        if keyword1 in title:
            print(datetime.now(),no,keyword1,title,url)
            url="http://"+url
            database.append([url])
        else:
            try:
                title2=browser.find_element_by_xpath('//*[@id="js-tracked_mod_5"]/div/section/div[1]/div/div/a[1]/h3').text
                url2=browser.find_element_by_xpath('//*[@id="js-tracked_mod_5"]/div/section/div[1]/div/div/a[1]/div/cite').text
                if keyword1 in title2:
                    print(datetime.now(),no,keyword1,title2,url2)
                    url2="http://"+url2
                    database.append([url2])
                else:
                    print(datetime.now(),no,keyword1,"error2:miss match","-")
                    database.append("-")
            except:
                print(datetime.now(),no,keyword1,"error2:miss match","-")
                database.append("-")
    except:
        print(datetime.now(),no,keyword1,"error3:no result","-")
        database.append("-")
        
    time.sleep(random_num(3))
    

df_out = pd.DataFrame(database,columns=["(*)PRTimes URL"])
df_concat=pd.concat( [df2,df_out], axis=1 )
df_concat.to_excel('sample_pr.xlsx', sheet_name='sheet1',index=False)
df_concat
```

```python
#キーワードからプレスリリースを検索
search_keyword_list=["映像伝送","XR技術"]

browser = webdriver.Chrome()
outbase=[]
    
for search_keyword in search_keyword_list:
    browser.get("https://prtimes.jp/")
    browser.find_element_by_xpath('/html/body/header/div/div[4]/div/input').send_keys(search_keyword)
    browser.find_element_by_xpath('/html/body/header/div/div[4]/div/input').send_keys(Keys.ENTER)

    ttl=browser.find_elements_by_xpath('/html/body/main/section/section/div/article[*]/a/div[1]/h3')
    urls=browser.find_elements_by_xpath('/html/body/main/section/section/div/article[*]/a')
    tm=browser.find_elements_by_xpath('/html/body/main/section/section/div/article[*]/a/div[1]/time')
    cn=browser.find_elements_by_xpath('/html/body/main/section/section/div/article[*]/span/a')



    for a,b,c,d in zip(ttl,urls,tm,cn):
        out=[search_keyword,d.text,c.text,a.text,b.get_attribute("href")]
        outbase.append(out)
    
    time.sleep(random_num(3))


df_ky=pd.DataFrame(outbase,columns=["カテゴリー","企業名","発表時間","記事タイトル","URL"])
df_ky.to_excel('キーワードからのプレスリリース.xlsx', sheet_name='sheet1',index=False)
df_ky
```
